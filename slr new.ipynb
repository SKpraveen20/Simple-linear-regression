```python
import pandas as pd
import numpy as np
import sklearn.metrics as mt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv("slr_data.csv")
x=data.iloc[:,0:1].values
y=data.iloc[:,1:2].values
```

```python
plt.scatter(x,y,color='orange')
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
data.describe()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
a={"actual":y_test.flatten(),"predicted":y_pred.flatten()}
df=pd.DataFrame(a)
df
```

```python
plt.scatter(x_train,y_train,color="c")
plt.plot(x_train,lr.predict(x_train),color="m")
plt.title("Simple Linear Regression on Train Set")
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
```

```python
plt.scatter(x_test,y_test,color="c")
plt.plot(x_train,lr.predict(x_train),color="m")
plt.title("Simple Linear Regression on Test Set")
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
print(lr.score(y_pred,y_test))
print('Root Mean Squared Error:', np.sqrt(mt.mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error:', np.sqrt(mt.mean_squared_error(y_train,lr.predict(x_train))))
```
